```python
import matplotlib.pyplot as plt
import numpy as np

# Criando dados para o gráfico
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Criando o gráfico
plt.plot(x, y, label="Seno")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Gráfico de uma Função Seno")
plt.legend()
plt.show()
```
